```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
```

```python
def make_soup(url: str) -> BeautifulSoup:
	source_code = requests.get(url, allow_redirects=False)
	plain_text = source_code.text.encode('ascii', 'replace')
	return BeautifulSoup(plain_text,'html.parser')
```

```python
soup = make_soup('http://ufcstats.com/fight-details/fe0c844a671dd6d8')
```

```python
tables = soup.findAll('tbody')
total_fight_data = [tables[0],tables[2]]
total_stats = []
for table in total_fight_data:
    row = table.find('tr')
    stats = ''
    for data in row.findAll('td'):
        if stats == '':
            stats = data.text
        else:    
            stats = stats + ',' + data.text
    total_stats.append(stats.replace('  ', '').replace('\n\n', '').replace('\n', ',').replace(', ', ',').replace(' ,',','))
    
total_stats[1] = ';'.join(total_stats[1].split(',')[6:])
total_stats[0] = ';'.join(total_stats[0].split(','))
total_stats = ';'.join(total_stats)
```

```python
header = 'R_fighter;B_fighter;R_KD;B_KD;R_SIG_STR.;B_SIG_STR.;R_SIG_STR_pct;B_SIG_STR_pct;R_TOTAL_STR.;B_TOTAL_STR.;R_TD;B_TD;R_TD_pct;B_TD_pct;R_SUB_ATT;B_SUB_ATT;R_PASS;B_PASS;R_REV;B_REV;R_HEAD;B_HEAD;R_BODY;B_BODY;R_LEG;B_LEG;R_DISTANCE;B_DISTANCE;R_CLINCH;B_CLINCH;R_GROUND;B_GROUND;win_by;last_round;last_round_time;Format;Referee;date;location;Fight_type;Winner'
```

```python
columns = ''
for div in soup.findAll('div', {'class':'b-fight-details__content'}):
    for col in div.findAll('p', {'class': 'b-fight-details__text'}):
        if columns == '':
            columns = col.text
        else:
            columns = columns + ',' +(col.text)

columns = columns.replace('  ', '').replace('\n\n\n\n', ',')\
.replace('\n', '').replace(', ', ',').replace(' ,',',')\
.replace('Method: ', '').replace('Round:', '').replace('Time:', '')\
.replace('Time format:', '').replace('Referee:', '')

fight_details = ';'.join(columns.split(',')[:5])
```

```python
winner = ''
for div in soup.findAll('div', {'class': 'b-fight-details__person'}):
    if div.find('i', {'class': 'b-fight-details__person-status b-fight-details__person-status_style_green'})!=None:
        winner = div.find('h3', {'class':'b-fight-details__person-name'}).text.replace(' \n', '').replace('\n', '')
```

```python
fight_type = soup.find("i",{"class":"b-fight-details__fight-title"}).text.replace('  ', '').replace('\n', '')
```

```python
another_soup = make_soup('http://ufcstats.com/event-details/e31502e3f79d00c5')
event_info = ''
for info in another_soup.findAll('li', {'class':'b-list__box-list-item'}):
    if event_info == '':
        event_info = (info.text)
    else:
        event_info = event_info + ';' + info.text 
    
event_info = ';'.join(event_info.replace('Date:','').replace('Location:','').replace('Attendance:','').replace('\n','').replace('  ', '').split(';')[:2])
```

```python
total_stats = total_stats + ';' + fight_details + ';' + event_info + ';' + fight_type + ';' + winner + '\n'
```
